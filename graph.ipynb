```python
import pandas as pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assign unique colors for each bar
colors = ['red', 'green','blue']
```

```python
albert = {
    "Untrained" : 0.5759,
    "Trained" : 0.7464,
    "Other domaintest set": 0.57600
}

# Extract keys and values
categories = list(albert.keys())
values = list(albert.values())

# Create the bar plot
plt.bar(categories, values, color=colors)

# Add labels and title
plt.xlabel("Stages")
plt.ylabel("Accuracy")
plt.title("Next Sentence Prediction Accuracy of ALBERT Model in different stages")

# Show the plot
plt.show()
```

```python
News = {
    "Albert" : 1.62005,
    "GPT-2" : 1.62733,
}

# Extract keys and values
categories = list(News.keys())
values = list(News.values())

# Create the bar plot
plt.bar(categories, values, color=colors)

# Add labels and title
plt.xlabel("Models")
plt.ylabel("Perplexity")
plt.title("Perplexity of Language Models in Trained domain test set")

# Set y-axis limit to zoom in and emphasize small differences
plt.ylim(min(values) - 0.01, max(values) + 0.01)

# Show the plot
plt.show()
```
